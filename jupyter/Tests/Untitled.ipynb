```python
import GPy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
modeWeights = pd.read_csv(r'Mode_Weight_Full_Cohort.csv',index_col='Mesh_ID')
print(modeWeights)
```

```python
modelOutput = pd.read_csv(r'simulation_output_Full_Cohort.csv',index_col=0)
print(modelOutput)
```

```python
modeMax=np.max(modeWeights,axis=0)
modeMin = np.min(modeWeights,axis=0)
modeNorm = (modeWeights-modeMin)/(modeMax-modeMin)
modeNorm = modeWeights/modeMax
modeNorm
```

```python
modeMin
```

```python
modeMax
```

```python
k = GPy.kern.RBF(18,ARD=True,name='rbf')
print(k)
```

```python
mGPy = GPy.models.GPRegression(modeNorm,np.array(modelOutput.iloc[:,0])[:,None],k )
```

```python
mGPy
```

```python
meanGP, covGP = mGPy.predict(np.array(modeNorm))
```

```python
np.mean(abs(meanGP-np.array(modelOutput.iloc[:,0])))
```

```python
mGPy.optimize()
mGPy
```
